## Post type and Average likes and Average shares and comments

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (replace with your dataset's file path)

facebook_data = pd.read_csv("Facebookdata.csv")

# Data cleaning: Convert relevant columns to numeric if needed
facebook_data['likes'] = pd.to_numeric(facebook_data['likes'], errors='coerce')
facebook_data['num_shares'] = pd.to_numeric(facebook_data['num_shares'], errors='coerce')
facebook_data['num_comments'] = pd.to_numeric(facebook_data['num_comments'], errors='coerce')

# Group by post_type and calculate the average likes, shares, and comments
engagement_by_post_type = facebook_data.groupby('post_type').agg({
    'likes': 'mean',
    'num_shares': 'mean',
    'num_comments': 'mean'
}).reset_index()

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot average likes
plt.bar(engagement_by_post_type['post_type'], engagement_by_post_type['likes'], 
        color='skyblue', label='Avg Likes', width=0.3, align='center')

# Plot average shares and comments as stacked bars
plt.bar(engagement_by_post_type['post_type'], engagement_by_post_type['num_shares'], 
        color='lightgreen', label='Avg Shares', width=0.3, align='edge')
plt.bar(engagement_by_post_type['post_type'], engagement_by_post_type['num_comments'], 
        color='lightcoral', label='Avg Comments', width=0.3, align='edge', bottom=engagement_by_post_type['num_shares'])

# Adding labels and title
plt.xlabel('Post Type')
plt.ylabel('Average Engagement')
plt.title('Average Likes, Shares, and Comments by Post Type')
plt.legend()
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()

# Display the engagement data for reference
engagement_by_post_type
```

## Engagement Analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your Facebook data into a DataFrame
# Replace 'your_file.csv' with the path to your dataset
facebook_data = pd.read_csv("Facebookdata.csv")

# Display the first few rows of the dataset
print(facebook_data.head())

# Ensure engagement columns are numeric
facebook_data['num_likes'] = pd.to_numeric(facebook_data['num_likes'], errors='coerce')
facebook_data['num_comments'] = pd.to_numeric(facebook_data['num_comments'], errors='coerce')
facebook_data['num_shares'] = pd.to_numeric(facebook_data['num_shares'], errors='coerce')

# Group by post type and calculate average engagement metrics
engagement_summary = facebook_data.groupby('post_type').agg({
    'num_likes': 'mean',
    'num_comments': 'mean',
    'num_shares': 'mean'
}).reset_index()

# Display the engagement summary
print(engagement_summary)

# Visualization
plt.figure(figsize=(12, 6))

# Bar plot for engagement metrics
engagement_summary.plot(x='post_type', kind='bar', legend=True)
plt.title('Average Engagement Metrics by Post Type')
plt.xlabel('Post Type')
plt.ylabel('Average Engagement')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()

# Optional: Additional analysis - t-test for differences in engagement
from scipy import stats

# Separate the engagement data by post type
original_posts = facebook_data[facebook_data['post_type'] == 'original']
shared_posts = facebook_data[facebook_data['post_type'] == 'shared']

# Perform t-tests
likes_ttest = stats.ttest_ind(original_posts['num_likes'].dropna(), shared_posts['num_likes'].dropna())
comments_ttest = stats.ttest_ind(original_posts['num_comments'].dropna(), shared_posts['num_comments'].dropna())
shares_ttest = stats.ttest_ind(original_posts['num_shares'].dropna(), shared_posts['num_shares'].dropna())

# Print t-test results
print(f"T-test for Likes: Statistic={likes_ttest.statistic}, p-value={likes_ttest.pvalue}")
print(f"T-test for Comments: Statistic={comments_ttest.statistic}, p-value={comments_ttest.pvalue}")
print(f"T-test for Shares: Statistic={shares_ttest.statistic}, p-value={shares_ttest.pvalue}")
```

## Post Type Sentiment Analysis

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from textblob import TextBlob

# Load the dataset

data = pd.read_csv("Facebookdata.csv")

# Post Type Analysis
# Group by 'post_type' and analyze engagement metrics
post_type_analysis = data.groupby('post_type').agg({
    'likes': 'mean',            # Use 'likes' instead of 'num_likes'
    'num_comments': 'mean',
    'num_shares': 'mean'
}).reset_index()

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x='post_type', y='likes', data=post_type_analysis, color='skyblue', label='Likes')
sns.barplot(x='post_type', y='num_comments', data=post_type_analysis, color='salmon', label='Comments')
sns.barplot(x='post_type', y='num_shares', data=post_type_analysis, color='lightgreen', label='Shares')
plt.legend()
plt.title('Average Engagement by Post Type')
plt.show()

# Content Sentiment Analysis
# Ensure there are no missing values in the 'content' column
data['content'].fillna('', inplace=True)

# Define a function to calculate sentiment
def get_sentiment(text):
    blob = TextBlob(text)
    return blob.sentiment.polarity

# Apply sentiment analysis on the content
data['content_sentiment'] = data['content'].apply(get_sentiment)

# Analyze the distribution of sentiment
plt.figure(figsize=(10, 6))
sns.histplot(data['content_sentiment'], kde=True, color='purple')
plt.title('Distribution of Post Sentiment')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()

# Analyze the relation between sentiment and engagement
sentiment_engagement = data.groupby(pd.cut(data['content_sentiment'], bins=[-1, -0.5, 0, 0.5, 1])).agg({
    'likes': 'mean',
    'num_comments': 'mean',
    'num_shares': 'mean'
}).reset_index()

# Plot sentiment vs engagement
plt.figure(figsize=(10, 6))
sns.lineplot(x='content_sentiment', y='likes', data=sentiment_engagement, marker='o', label='Likes')
sns.lineplot(x='content_sentiment', y='num_comments', data=sentiment_engagement, marker='o', label='Comments')
sns.lineplot(x='content_sentiment', y='num_shares', data=sentiment_engagement, marker='o', label='Shares')
plt.title('Engagement vs Sentiment')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Average Engagement')
plt.legend()
plt.show()
```

## Sponsorship and ENgagement Analysis

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset

data = pd.read_csv("Facebookdata.csv")

# Sponsorship and Engagement Analysis
# Fill missing values in relevant columns
data['video_view_count'].fillna(0, inplace=True)
data['likes'].fillna(0, inplace=True)
data['num_comments'].fillna(0, inplace=True)
data['num_shares'].fillna(0, inplace=True)

# Convert 'is_sponsored' to numeric if needed (0 for False, 1 for True)
data['is_sponsored'] = data['is_sponsored'].astype(int)

# Calculate average engagement metrics for sponsored vs non-sponsored posts
sponsorship_engagement = data.groupby('is_sponsored').agg({
    'likes': 'mean',
    'num_comments': 'mean',
    'num_shares': 'mean',
    'video_view_count': 'mean'
}).reset_index()

# Rename the 'is_sponsored' column for clarity in the plot
sponsorship_engagement['is_sponsored'] = sponsorship_engagement['is_sponsored'].replace({0: 'Non-Sponsored', 1: 'Sponsored'})

# Plot the average engagement metrics for sponsored vs non-sponsored posts
plt.figure(figsize=(12, 6))

# Create barplots for each engagement metric
plt.subplot(1, 3, 1)
sns.barplot(x='is_sponsored', y='likes', data=sponsorship_engagement, palette='Blues')
plt.title('Average Likes by Sponsorship')

plt.subplot(1, 3, 2)
sns.barplot(x='is_sponsored', y='num_comments', data=sponsorship_engagement, palette='Oranges')
plt.title('Average Comments by Sponsorship')

plt.subplot(1, 3, 3)
sns.barplot(x='is_sponsored', y='num_shares', data=sponsorship_engagement, palette='Greens')
plt.title('Average Shares by Sponsorship')

plt.tight_layout()
plt.show()

# Video Performance and Sponsorship
# Analyze the impact of video views based on whether a post is sponsored or not
plt.figure(figsize=(8, 6))
sns.barplot(x='is_sponsored', y='video_view_count', data=sponsorship_engagement, palette='Purples')
plt.title('Average Video Views by Sponsorship')
plt.xlabel('Sponsorship Status')
plt.ylabel('Average Video Views')
plt.show()

# Engagement vs Video View Count
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_view_count', y='likes', hue='is_sponsored', data=data, palette='coolwarm')
plt.title('Likes vs Video Views (Sponsored vs Non-Sponsored)')
plt.xlabel('Video View Count')
plt.ylabel('Likes')
plt.legend(title='Sponsorship')
plt.show()

# Similar scatterplot for Comments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_view_count', y='num_comments', hue='is_sponsored', data=data, palette='coolwarm')
plt.title('Comments vs Video Views (Sponsored vs Non-Sponsored)')
plt.xlabel('Video View Count')
plt.ylabel('Comments')
plt.legend(title='Sponsorship')
plt.show()

# Similar scatterplot for Shares
plt.figure(figsize=(10, 6))
sns.scatterplot(x='video_view_count', y='num_shares', hue='is_sponsored', data=data, palette='coolwarm')
plt.title('Shares vs Video Views (Sponsored vs Non-Sponsored)')
plt.xlabel('Video View Count')
plt.ylabel('Shares')
plt.legend(title='Sponsorship')
plt.show()
```

## Average Media Usage Analysis

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("Facebookdata.csv")

# Fill missing values in relevant columns
data['likes'].fillna(0, inplace=True)
data['num_comments'].fillna(0, inplace=True)
data['num_shares'].fillna(0, inplace=True)

# Post Type Analysis
# Group by 'post_type' and calculate the average engagement metrics
post_type_engagement = data.groupby('post_type').agg({
    'likes': 'mean',
    'num_comments': 'mean',
    'num_shares': 'mean'
}).reset_index()

# Plot the average engagement metrics for different post types
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.barplot(x='post_type', y='likes', data=post_type_engagement, palette='Blues')
plt.title('Average Likes by Post Type')

plt.subplot(1, 3, 2)
sns.barplot(x='post_type', y='num_comments', data=post_type_engagement, palette='Oranges')
plt.title('Average Comments by Post Type')

plt.subplot(1, 3, 3)
sns.barplot(x='post_type', y='num_shares', data=post_type_engagement, palette='Greens')
plt.title('Average Shares by Post Type')

plt.tight_layout()
plt.show()

# Media Usage Analysis (Images and External Links)
# Create a new column to check if a post has media (post_image, post_external_image)
data['has_media'] = data.apply(lambda x: 1 if pd.notna(x['post_image']) or pd.notna(x['post_external_image']) else 0, axis=1)

# Group by 'has_media' (1 = has media, 0 = no media) and calculate engagement metrics
media_engagement = data.groupby('has_media').agg({
    'likes': 'mean',
    'num_comments': 'mean',
    'num_shares': 'mean'
}).reset_index()

# Rename 'has_media' column for clarity in the plot
media_engagement['has_media'] = media_engagement['has_media'].replace({0: 'No Media', 1: 'Has Media'})

# Plot the average engagement metrics for posts with and without media
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.barplot(x='has_media', y='likes', data=media_engagement, palette='Purples')
plt.title('Average Likes by Media Usage')

plt.subplot(1, 3, 2)
sns.barplot(x='has_media', y='num_comments', data=media_engagement, palette='Reds')
plt.title('Average Comments by Media Usage')

plt.subplot(1, 3, 3)
sns.barplot(x='has_media', y='num_shares', data=media_engagement, palette='Greens')
plt.title('Average Shares by Media Usage')

plt.tight_layout()
plt.show()

# Scatterplot to explore relation between post type and engagement
plt.figure(figsize=(10, 6))
sns.scatterplot(x='post_type', y='likes', hue='has_media', data=data, palette='coolwarm')
plt.title('Likes by Post Type and Media Usage')
plt.xlabel('Post Type')
plt.ylabel('Likes')
plt.legend(title='Media Usage')
plt.show()

# Scatterplot for comments
plt.figure(figsize=(10, 6))
sns.scatterplot(x='post_type', y='num_comments', hue='has_media', data=data, palette='coolwarm')
plt.title('Comments by Post Type and Media Usage')
plt.xlabel('Post Type')
plt.ylabel('Comments')
plt.legend(title='Media Usage')
plt.show()

# Scatterplot for shares
plt.figure(figsize=(10, 6))
sns.scatterplot(x='post_type', y='num_shares', hue='has_media', data=data, palette='coolwarm')
plt.title('Shares by Post Type and Media Usage')
plt.xlabel('Post Type')
plt.ylabel('Shares')
plt.legend(title='Media Usage')
plt.show()
```

## Price Distribution Analysis 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('Facebook Marketplace.csv')

# Convert final_price to numeric, if it's not already
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

# Drop rows with NaN in final_price
df = df.dropna(subset=['final_price'])

# Price Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['final_price'], bins=30, kde=True)
plt.title('Price Distribution of Products')
plt.xlabel('Final Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## Category Analysis

```python
# Category Analysis
plt.figure(figsize=(12, 6))
category_counts = df['root_category'].value_counts()
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Number of Products by Category')
plt.xlabel('Category')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## Seller Popularity Analysis

```python
# Calculate average final price per seller
average_price_per_seller = df.groupby('seller_description')['final_price'].mean().sort_values(ascending=False)

# Plotting Seller Popularity
plt.figure(figsize=(12, 6))
sns.barplot(x=average_price_per_seller.index[:10], y=average_price_per_seller.values[:10])  # Top 10 sellers
plt.title('Top 10 Sellers by Average Final Price of Products')
plt.xlabel('Seller')
plt.ylabel('Average Final Price')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

## Location Based Analysis

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Facebook Marketplace.csv')

# Display the first few rows of the DataFrame
print(df.head())

# Basic Information about the DataFrame
print(df.info())

# Analysis: Count of listings per location
location_counts = df['location'].value_counts()

# Display the top 10 locations with the most listings
print("Top 10 Locations with Most Listings:")
print(location_counts.head(10))

# Analysis: Average final price per location
average_price_per_location = df.groupby('location')['final_price'].mean().sort_values(ascending=False)

# Display the average final price per location
print("Average Final Price per Location:")
print(average_price_per_location.head(10))

# Visualizing the results using Matplotlib (optional)
import matplotlib.pyplot as plt

# Plotting the number of listings per location
plt.figure(figsize=(12, 6))
location_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Locations with Most Listings')
plt.xlabel('Location')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.show()

# Plotting the average final price per location
plt.figure(figsize=(12, 6))
average_price_per_location.head(10).plot(kind='bar', color='lightgreen')
plt.title('Average Final Price per Location')
plt.xlabel('Location')
plt.ylabel('Average Final Price')
plt.xticks(rotation=45)
plt.show()
```
